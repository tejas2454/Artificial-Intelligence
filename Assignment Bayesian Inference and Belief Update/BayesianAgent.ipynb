# Belief

```python
# prob_W_s : Probability of Weather = Sunny
# prob_W_c : Probability of Weather = Cloudy

# prob_P_a : Probability of Partner = Available
# prob_P_n : Probability of Partner = Not Available

# prob_M   : Probability of Movie == True
# prob_T   : Probability of Tennis == True

# Prob_M_g_s_n : Probability of Movie given Weather : Sunny and Person : Not Available
# Prob_M_g_s_a : Probability of Movie given Weather : Sunny and Person : Available
# Prob_M_g_c_n : Probability of Movie given Weather : Cloudy and Person : Not Available
# Prob_M_g_c_a : Probability of Movie given Weather : Cloudy and Person : Available

# Prob_T_g_s_n : Probability of Tennis given Weather : Sunny and Person : Not Available
# Prob_T_g_s_a : Probability of Tennis given Weather : Sunny and Person : Available
# Prob_T_g_c_n : Probability of Tennis given Weather : Cloudy and Person : Not Available
# Prob_T_g_c_a : Probability of Tennis given Weather : Cloudy and Person : Available

print('===========================   Belief   ===========================')

prob_W_s = 0.7
prob_W_c = 0.3 

prob_P_a = 0.5
prob_P_n = 0.5 

prob_M_g_s_a = 0.01
prob_M_g_s_n = 0.60
prob_M_g_c_a = 0.10
prob_M_g_c_n = 0.90

prob_T_g_s_a = 0.95
prob_T_g_s_n = 0.10
prob_T_g_c_a = 0.10
prob_T_g_c_n = 0.01

# Formule Used
# prob_M = prob_M_g_s_a*prob_W_s(intersection)prob_P_a + prob_M_g_s_n*prob_W_s(intersection)prob_P_n + prob_M_g_c_a*prob_W_c(intersection)prob_P_a + prob_M_g_c_n*prob_W_c(intersection)prob_P_n
# prob_W_c(intersection)prob_P_n = prob_W_c * prob_P_n, as both the events are independent of each other


prob_M = prob_M_g_s_a*prob_W_s*prob_P_a + prob_M_g_s_n*prob_W_s*prob_P_n + prob_M_g_c_a*prob_W_c*prob_P_a + prob_M_g_c_n*prob_W_c*prob_P_n
 
prob_T = prob_T_g_s_a*prob_W_s*prob_P_a + prob_T_g_s_n*prob_W_s*prob_P_n + prob_T_g_c_a*prob_W_c*prob_P_a + prob_T_g_c_n*prob_W_c*prob_P_n

print()
print('Probability of Weather = Sunny: {}'.format(prob_W_s))
print('Probability of Weather = Cloudy:{}'.format(prob_W_c))
print()
print('Probability of Partner = Available:{}'.format(prob_P_a))
print('Probability of Partner = Not Available:{}'.format(prob_P_n))
print()
print('Probability of Movie given Weather : Sunny and Person : Available:{}'.format(prob_M_g_s_a))
print('Probability of Movie given Weather : Sunny and Person : Not Available : {}'.format(prob_M_g_s_n))
print('Probability of Movie given Weather : Cloudy and Person : Available : {}'.format(prob_M_g_c_a))
print('Probability of Movie given Weather : Cloudy and Person : Not Available :{}'.format(prob_M_g_c_n))
print()
print('Probability of Tennis given Weather : Sunny and Person : Available:{}'.format(prob_T_g_s_a))
print('Probability of Tennis given Weather : Sunny and Person : Not Available :{}'.format(prob_T_g_s_n))
print('Probability of Tennis given Weather : Cloudy and Person : Available :{}'.format(prob_T_g_c_a))
print('Probability of Tennis given Weather : Cloudy and Person : Not Available :{}'.format(prob_T_g_c_n))
print()
print('Calculated Probability of Movie == True:{}'.format(np.round(prob_M,3)))
print('Calculated Probability of Tennis == True:{}'.format(np.round(prob_T,3)))
print()
```

```python
if prob_M > 0.3 and prob_T < 0.3:
    print('Desire : Movie with probability : {}'.format(np.round(prob_M,3)))
    intension = 'Movie'
elif prob_M > 0.3 and prob_T > 0.3:
    print('Desire : Movie and Tennis with probability : {} and {} respectively'.format(np.round(prob_M,3),np.round(prob_T,3)))
    if prob_M > prob_T:
        print('Most Probable desire is Movie') # In case I am designing the agent, I would have implemented this condition, as there should a favorable condition
        intension = 'Movie'
    else:
        print('Most Probable desire is Tennis')    
        intension = 'Tennis'
elif prob_M < 0.3 and prob_T > 0.3:
    print('Desire : Tennis with probability : {}'.format(np.round(prob_T,3)))
    intension = 'Tennis'
elif prob_M < 0.3 and prob_T < 0.3:
    print('Desire : None ')
    if prob_M < prob_T:
        print('Least Probable desire is Movie') # In case I am designing the agent, I would have implemented this condition, as there should a favorable condition
        intension = 'Tennis'
    else:
        print('Least Probable desire is Tennis') 
        intension = 'Movie'
            
```
